```python
import numpy as np
import matplotlib.pyplot as plt

# Datos ficticios para representar el tiempo de ejecucin en funcin del tamao del primo
bits = np.array([12, 14, 16, 18, 20, 22, 24, 26, 28])

# Tiempo de ejecucin con raz primitiva (ms seguro)
time_primitive = np.array([0.0001, 0.0003, 0.0007, 0.0015, 0.003, 0.007, 0.015, 0.025, 0.035])

# Tiempo de ejecucin con un generador no primitivo (menos seguro)
time_non_primitive = np.array([0.00005, 0.0002, 0.0006, 0.0012, 0.0028, 0.006, 0.012, 0.020, 0.028])

# Crear la grfica
plt.figure(figsize=(8, 5))
plt.plot(bits, time_primitive, 'o-', label="BSGS (Raz Primitiva)", color='blue')
plt.plot(bits, time_non_primitive, 's-', label="BSGS (No Primitiva)", color='orange')

# Etiquetas y ttulo
plt.xlabel("Tamao del nmero primo (bits)")
plt.ylabel("Tiempo de ejecucin (s)")
plt.title("BSGS: Raz Primitiva vs No Primitiva")
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos ficticios para representar el tiempo de ejecucin con Brute Force y BSGS
bits = np.array([12, 14, 16, 18, 20, 22, 24, 26, 28])

# Tiempo de ejecucin con Brute Force (escala logartmica)
time_brute_force = np.array([1e-3, 5e-2, 1e-1, 3e-1, 1, 3, 10, 30, 100])

# Tiempo de ejecucin con BSGS (escala logartmica)
time_bsgs = np.array([1e-4, 2e-3, 5e-3, 1e-2, 3e-2, 8e-2, 2e-1, 5e-1, 2])

# Crear la grfica
plt.figure(figsize=(8, 5))
plt.plot(bits, time_brute_force, 'o-', label="Brute Force", color='blue')
plt.plot(bits, time_bsgs, 's-', label="Baby-step Giant-step (BSGS)", color='orange')

# Escala logartmica en el eje Y
plt.yscale("log")

# Etiquetas y ttulo
plt.xlabel("Tamao del nmero primo (bits)")
plt.ylabel("Tiempo de ejecucin (s)")
plt.title("Comparacin de Mtodos para Resolver Logaritmo Discreto")
plt.legend()
plt.grid(True, which="both", linestyle="--")

# Mostrar la grfica
plt.show()
```
