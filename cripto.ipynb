```python
import numpy as np
import matplotlib.pyplot as plt

# Datos ficticios para representar el tiempo de ejecucin en funcin del tamao del primo
bits = np.array([12, 14, 16, 18, 20, 22, 24, 26, 28])

# Tiempo de ejecucin con raz primitiva (ms seguro)
time_primitive = np.array([0.0001, 0.0003, 0.0007, 0.0015, 0.003, 0.007, 0.015, 0.025, 0.035])

# Tiempo de ejecucin con un generador no primitivo (menos seguro)
time_non_primitive = np.array([0.00005, 0.0002, 0.0006, 0.0012, 0.0028, 0.006, 0.012, 0.020, 0.028])

# Crear la grfica
plt.figure(figsize=(8, 5))
plt.plot(bits, time_primitive, 'o-', label="BSGS (Raz Primitiva)", color='blue')
plt.plot(bits, time_non_primitive, 's-', label="BSGS (No Primitiva)", color='orange')

# Etiquetas y ttulo
plt.xlabel("Tamao del nmero primo (bits)")
plt.ylabel("Tiempo de ejecucin (s)")
plt.title("BSGS: Raz Primitiva vs No Primitiva")
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()
```
